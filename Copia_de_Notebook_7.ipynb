14) ¬øCual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¬øQue paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy
#maxima = np.array(expvida['life_expectancy'])
#maxim_exp = np.max(array)
```

```python
# Expectativa de vida m√°xima y m√≠nima usando pandas
max_exp_pandas = expvida['life_expectancy'].max()
min_exp_pandas = expvida['life_expectancy'].min()

# Pa√≠ses con la expectativa de vida m√°s alta y m√°s baja
pais_max = expvida.loc[expvida['life_expectancy'] == max_exp_pandas, 'Country'].values[0]
pais_min = expvida.loc[expvida['life_expectancy'] == min_exp_pandas, 'Country'].values[0]

print(f"M√°xima expectativa de vida (pandas): {max_exp_pandas} a√±os en {pais_max}")
print(f"M√≠nima expectativa de vida (pandas): {min_exp_pandas} a√±os en {pais_min}")
```

```python
import numpy as np

# Convertir la columna 'life_expectancy' a un array de NumPy
maxima = np.array(expvida['life_expectancy'])

# M√°xima y m√≠nima expectativa de vida usando NumPy
maxim_exp_numpy = np.max(maxima)
minim_exp_numpy = np.min(maxima)

# Verificar que el valor m√°ximo y m√≠nimo realmente exista
pais_max_numpy = expvida.loc[expvida['life_expectancy'] == maxim_exp_numpy, 'Country']
pais_min_numpy = expvida.loc[expvida['life_expectancy'] == minim_exp_numpy, 'Country']

# Asegurarnos de que los resultados no est√°n vac√≠os
if not pais_max_numpy.empty:
    pais_max_numpy = pais_max_numpy.values[0]
else:
    pais_max_numpy = 'No encontrado'

if not pais_min_numpy.empty:
    pais_min_numpy = pais_min_numpy.values[0]
else:
    pais_min_numpy = 'No encontrado'

print(f"M√°xima expectativa de vida (NumPy): {maxim_exp_numpy} a√±os en {pais_max_numpy}")
print(f"M√≠nima expectativa de vida (NumPy): {minim_exp_numpy} a√±os en {pais_min_numpy}")

```

18) ¬øCuantos paises presentaron algun caso de Sarampion (**Measles**) en el 2015? **_Pista_**: Lee acerca de la funci√≥n [.count()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html?highlight=count#pandas.DataFrame.count)

```python
# Filtramos el DataFrame para el a√±o 2015
# Verificar los nombres de las columnas
print(expvida.columns)
# Filtrar el DataFrame para el a√±o 2015
paisesSarampion = expvida[(expvida['Year'] == 2015)]

# Filtrar los pa√≠ses con casos de sarampi√≥n (measles)
paisesConSarampion = paisesSarampion[paisesSarampion['measles'] > 0]

# Contar el n√∫mero de pa√≠ses con casos de sarampi√≥n
num_paises_sarampion = paisesConSarampion['Country'].nunique()

print(f"El n√∫mero de pa√≠ses que tuvieron casos de sarampi√≥n en 2015 es: {num_paises_sarampion}")

```
